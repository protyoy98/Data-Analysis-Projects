--


Now we have to extract some new features from started_at column like ride_year, ride_month, ride_day_of_week, ride_week_no and ride_hour


--

```python
df_all_3['ride_year'] = df_all_3.started_at.dt.year
df_all_3['ride_month'] = df_all_3.started_at.dt.month_name()
df_all_3['ride_day_of_week'] = df_all_3.started_at.dt.day_name() #there will be an extra column for this...we shall drop this later
df_all_3['ride_week_no'] = df_all_3.started_at.dt.weekofyear
df_all_3['ride_hour'] = df_all_3.started_at.dt.hour
```

```python
#checking each new column has the correct values
for col in ['ride_year','ride_month','ride_day_of_week','ride_week_no','ride_hour']:
    print(col, df_all_3[col].unique())
    print(col, df_all_3[col].unique().shape)
```

Now we shall create a weekday_weekend column from ride_day_of_week

--

We shall create a season column from ride_month as given in this article - https://www.newyork.com.au/travel-season-new-york/

--

```python
# first, we shall create a function
def get_season(mon_name):
        if mon_name in ['December','January','February']:
            return 'Winter(Dec-Feb)'
        elif mon_name in ['March','April','May']:
            return 'Spring(Mar-May)'
        elif mon_name in ['June','July','August']:
            return 'Summer(Jun-Aug)'
        elif mon_name in ['September','October','November']:
            return 'Autumn(Sep-Nov)'
```

```python
# Now, lets apply the function to ride_month column
df_all_3['ride_season'] = df_all_3.ride_month.apply(get_season)
```

```python
#verify
df_all_3.sample(5)
```

```python
df_all_3.ride_season.unique()
```

--


Looks good till now. Now lets calculate ride_len from started_at and ended_at columns


--

```python
#creating a new duplicate dataframe
df_all_4 = df_all_3
```

```python
#assigning a variable and then creating a new column
ride_len = df_all_4.ended_at - df_all_4.started_at
```

```python
ride_len = np.round(ride_len.dt.total_seconds() / 60, 2)
```

```python
df_all_4['ride_len_mins'] = ride_len
```

```python
#verify
df_all_4.sample(5)
```

```python
#lets look for any invalid or outliers in ride_len_mins
df_all_4.describe()
```

--

Mmm....looks like there are some negative values as well as some outliers in our data. 41629.17 mins = 29 days. This can be an actual datapoint but we shall treat this as outliers and remove this from dataframe for analysis purposes.

--

```python
# First, let's deal with negative values

df_all_4[df_all_4.ride_len_mins < 0]
```

```python
# We shall be dropping these values

df_all_4 = df_all_4[df_all_4.ride_len_mins >= 0]
df_all_4.describe()
```

```python
# Now let's deal with outliers

mean_ride_len = df_all_4.ride_len_mins.mean()
std_dev_ride_len = df_all_4.ride_len_mins.std()
upper_limit = mean_ride_len + 3 * std_dev_ride_len
lower_limit = 0
```

```python
#first lets see the ouliers


df_all_4[(df_all_4.ride_len_mins > upper_limit) | (df_all_4.ride_len_mins < lower_limit)]
```

```python
# As the outliers count for only a small percentage of data and imputation is not possible, we shall drop these records

df_all_4 = df_all_4[(df_all_4.ride_len_mins <= upper_limit) & (df_all_4.ride_len_mins >= lower_limit)]
df_all_4.describe()
```
